# Time Complexity

## Runtime efficiency - Loops

```python
nList=[10,50,100,200,500,750,1000,2000,3000,4000,5000,7000,10000,20000,30000,40000,50000]
timeList=[]

for j in range(len(nList)):
    n = nList[j]
    a = ['NA']*n
    start = time.time()
    for i in range(n):
        a[i] = 0
    end = time.time()
    time_diff = end - start
    timeList.append(time_diff)
    
```

```python
plt.plot(nList,timeList)
plt.ylabel('time')
axes = plt.gca()
plt.show()
```

## Runtime efficiency - Calculating mean

#### Without recursion

```python
nList=[10,50,100,200,500,750,1000,2000,3000,4000,5000,7000,10000,20000,30000,40000,50000]
timeList=[]
for k in range(len(nList)):
    n = nList[k]
    x = np.arange(1,n+1)
    avg=0
    avg=[0]*n
    start = time.time()
    for i in range(n):
        s = 0
    
        for j in range(i+1):
            s=s+x[j]
            avg[i] = np.float(s)/(i+1)
            
    end = time.time()
    time_diff = end - start
    timeList.append(time_diff)
```

#### with recursion

```python
nList=[10,50,100,200,500,750,1000,2000,3000,4000,5000,7000,10000,20000,30000,40000,50000]
timeList_rec=[]
for k in range(len(nList)):
    n = nList[k]
    x = np.arange(1,n+1)
    avg=[0]*n
    avg[0] = x[0]
    start = time.time()
    for i in range(1,n):
        avg[i] = np.float((avg[i-1]*i+x[i]))/(i+1)
    end = time.time()
    time_diff = end - start
    timeList_rec.append(time_diff)

        

       
```

```python
plt.plot(nList,timeList_rec)
plt.ylabel('time')
axes = plt.gca()
plt.show()
```

```python
plt.plot(nList,timeList_rec)
plt.plot(nList,timeList)
plt.ylabel('time')
axes = plt.gca()
plt.show()
```
